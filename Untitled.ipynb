```python
chunk.head()
```

```python
# Troca 2 (eliminado na prova) por 0. Entao 0 passa a representar Faltantes e eliminados
chunk.loc[chunk.TP_PRESENCA_CN == 2, 'TP_PRESENCA_CN'] = 0
chunk.loc[chunk.TP_PRESENCA_CH == 2, 'TP_PRESENCA_CH'] = 0
chunk.loc[chunk.TP_PRESENCA_LC == 2, 'TP_PRESENCA_LC'] = 0
chunk.loc[chunk.TP_PRESENCA_MT == 2, 'TP_PRESENCA_MT'] = 0

# mudando as colunas para o tipo categorico
chunk["TP_PRESENCA_CN"] = chunk["TP_PRESENCA_CN"].astype('category')
chunk["TP_PRESENCA_CH"] = chunk["TP_PRESENCA_CH"].astype('category')
chunk["TP_PRESENCA_LC"] = chunk["TP_PRESENCA_LC"].astype('category')
chunk["TP_PRESENCA_MT"] = chunk["TP_PRESENCA_MT"].astype('category')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
f, axes = plt.subplots(2, 2)
f.set_size_inches(18.5, 10.5, forward=True)

sns.set(style="whitegrid")

sns.countplot(x="TP_PRESENCA_CN", data=chunk, ax=axes[0][0])
sns.countplot(x="TP_PRESENCA_CH", data=chunk, ax=axes[0][1])
sns.countplot(x="TP_PRESENCA_LC", data=chunk, ax=axes[1][0])
sns.countplot(x="TP_PRESENCA_MT", data=chunk, ax=axes[1][1])
```

```python
chunk['PRESENCA'] = chunk.apply(lambda row: row.TP_PRESENCA_CN and
                                            row.TP_PRESENCA_CH and
                                            row.TP_PRESENCA_LC and
                                            row.TP_PRESENCA_MT, axis=1)
```
